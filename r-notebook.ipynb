```r
# Read data from CSV file
data <- read.csv("/kaggle/input/r-data/r_pie_data.csv")

png("pie_chart.png", width = 800, height = 600)  # Change width and height as needed

pie(data$Value,
    col = rainbow(nrow(data)),
    main = "Pie Chart",
    cex = 0.8,
    labels = paste0(data$Value))  

legend("right", legend = data$Category, fill = rainbow(nrow(data)))
title(main = "Motives of Cyber Crime", line = -2)
mtext("2020 Data from data.gov.in", side = 1, line = 0)
dev.off() 
```
